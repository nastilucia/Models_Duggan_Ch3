```julia
using Plots

#parameters

ref_gr = 0.1 #Reference Growth Rate
ref_a = 1.0 #Effect of Availability on Growth Rate
availability = [1.0, 0.5, 0.0]
growth_rate_arr = zeros(size(availability))

for i in 1:3
    eff_a = availability[i]/ref_a
    growth_rate_arr[i] = ref_gr * eff_a
end

lineplot = plot(availability,growth_rate_arr)
display(lineplot)
```
